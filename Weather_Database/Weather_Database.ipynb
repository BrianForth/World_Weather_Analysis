```python
# Import dependencies
import pandas as pd
import numpy as np
from citipy import  citipy
import time
# Import requests library
import requests
# Import API key
from config import weather_api_key
# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
# Create random lat-lon pairs
lat = np.random.uniform(-90.000, 90.000, size=2000)
lon = np.random.uniform(-180.000, 180.000, size=2000)
lat_lon = zip(lat, lon)
lat_lon
```

```python
# Add lat-lon pairs to list
coordinates = list(lat_lon)
```

```python
# Create list to hold cities
cities = []
# ID nearest city to each lat-long pair
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    # Add to list if unique
    if city not in cities:
        cities.append(city)
# Print city count to confirm sufficient length
len(cities)
```

```python
# Create empty list to hold weather data
city_data = []
# Print log start
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters
record_count = 1
set_count = 1

# Loop through all cities in list
for i, city in enumerate(cities):
    
    # Group cities by 50 for logging
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
        time.sleep(40)
    # Create endpoint url for each city
    city_url = url + "&q=" + city.replace(" ", "+")
    
    # Log url, record and set numbers, and city name
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    # Iterate record count
    record_count += 1
    # Run API request for city
    try: 
        # Parse JSON and retrieve data
        city_weather = requests.get(city_url).json()
        # Parse out data of interest
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_overall = city_weather["weather"][0]["description"]
        
        # Append city info into city_data
        city_data.append({"City": city.title(),
                         "Country": city_country,
                         "Lat": city_lat,
                         "Lng": city_lng,
                         "Max Temp": city_max_temp,
                         "Humidity": city_humidity,
                         "Cloudiness": city_clouds,
                         "Wind Speed": city_wind,
                         "Current Description": city_overall})
    # Skip city if error occurs
    except (KeyError):
        print("City not found. Skipping...")
        pass
# Indicate that data loading is complete
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
# Create output file (CSV)
output_data_file = "WeatherPy_Database.csv"
# Export DF into CSV
city_data_df.to_csv(output_data_file, index_label="City_ID")
```
